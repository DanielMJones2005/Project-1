```python
# Dependencies
import requests
import csv
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# import api_key from config file
from config import eia_key
from pprint import pprint
```

# CO2 Emission by State - Annual

```python
series_url = "http://api.eia.gov/series/?" + "api_key=" + eia_key + "&series_id="

state_abbr = ["AL", "AK", "AZ", "AR", "CA", "CO", "CT", "DE", "FL", "GA",
           "HI", "ID", "IL", "IN", "IA", "KS", "KY", "LA", "ME", "MD",
           "MA", "MI", "MN", "MS", "MO", "MT", "NE", "NV", "NH", "NJ",
           "NM", "NY", "NC", "ND", "OH", "OK", "OR", "PA", "RI", "SC",
           "SD", "TN", "TX", "UT", "VT", "VA", "WA", "WV", "WI", "WY"]
emission_state_pre = "EMISS.CO2-TOTV-IC-TO-"
emission_state_post = ".A"

header_df = pd.DataFrame()
# # set up additional columns to hold header information
header_df["pk_id"] = ""
header_df["sector"] = ""
header_df["energy_source"] = ""
header_df["frequency"] = ""
header_df["name"] = ""
header_df["source"] = ""
header_df["series_id"] = ""
header_df["unit_long"] = ""
header_df["unit_short"] = ""
header_df["state"] = ""
header_df["data_for"] = ""

line_df = pd.DataFrame()
# # set up additional columns to hold line information
line_df["fk_id"] = ""
line_df["year"] = ""
line_df["month"] = ""
line_df["value"] = ""


for count, st in enumerate(state_abbr, start=1):
    query_url = series_url + emission_state_pre + st + emission_state_post
    st_co2_json = requests.get(query_url).json()
    header_df = header_df.append({"pk_id" : count,
                                  "sector" : st_co2_json["series"][0]['name'].split(" ")[0],
                                  "name" : st_co2_json["series"][0]['name'],
                                  "source" : st_co2_json["series"][0]['source'],
                                  "frequency" : "Annual",
                                  "series_id" : st_co2_json["series"][0]['series_id'], 
                                  "unit_long" : st_co2_json["series"][0]['units'],
                                  "unit_short" : st_co2_json["series"][0]['unitsshort'],
                                  "state" : st_co2_json["series"][0]['geography'].split("-")[1],
                                  "data_for" : "CO2-Emission-By-State",
                                 } , ignore_index=True )
    
    for dt in st_co2_json["series"][0]["data"]:
            line_df = line_df.append({"fk_id" : count,
                              "year" : dt[0],
                              "month" : 13,
                              "value" : dt[1]        
                               } , ignore_index=True )
    pkid = count
    
print("-----------------------------")
print("Data Retrieval Complete for CO2 Emission by State")
print("-----------------------------")
    
```

# CO2 Emission - Annual

```python
emission_pre = "TOTAL."
emission_post = ".A"
energy_source = ["CLICEUS", "DFICEUS", "HLICEUS", "KSICEUS", 
                "LUICEUS", "MMICEUS", "NNICEUS", "OPICEUS",
                "PCICEUS", "PMICEUS", "RFICEUS", "TEICEUS"]

for count, es in enumerate(energy_source, start=pkid+1):
    query_url = series_url + emission_pre + es + emission_post
    co2_json = requests.get(query_url).json()
    header_df = header_df.append({"pk_id" : count,
                                  "sector" : "Industrial",
                                  "energy_source" : co2_json["series"][0]['name'].split(" ")[0],
                                  "frequency" : "Annual",
                                  "name" : co2_json["series"][0]['name'],
                                  "source" : "EIA, U.S. Energy Information Administration",
                                  "series_id" : co2_json["series"][0]['series_id'], 
                                  "unit_long" : co2_json["series"][0]['units'],
                                  "data_for" : "CO2-Emission",
                                 } , ignore_index=True )
    
    for dt in co2_json["series"][0]["data"]:
            line_df = line_df.append({"fk_id" : count,
                                      "year" : dt[0],
                                      "month" : 13,
                                      "value" : dt[1]
                               } , ignore_index=True )
    pkid = count
    
print("-----------------------------")
print("Data Retrieval Complete for CO2 Emission")
print("-----------------------------")
```

# Energy Consumption - Annual

```python
consumption_pre = "TOTAL."
consumption_post = ".A"
consumption_source = ["TEICBUS", "WYICBUS", "CLICBUS", "HVICBUS",
                      "GEICBUS", "NNICBUS", "PMICBUS", "SOICBUS"]

for count, cs in enumerate(consumption_source, start=pkid+1):
    query_url = series_url + consumption_pre + cs + consumption_post
    cons_json = requests.get(query_url).json()
    header_df = header_df.append({"pk_id" : count,
                                  "sector" : "Industrial",
                                  "energy_source" : cons_json["series"][0]['name'].split(" ")[0],
                                  "frequency" : "Annual",
                                  "name" : cons_json["series"][0]['name'],
                                  "source" : "EIA, U.S. Energy Information Administration",
                                  "series_id" : cons_json["series"][0]['series_id'], 
                                  "unit_long" : cons_json["series"][0]['units'],
                                  "data_for" : "Consumption",
                                 } , ignore_index=True )
    
    for dt in cons_json["series"][0]["data"]:
            line_df = line_df.append({"fk_id" : count,
                                      "year" : dt[0],
                                      "month" : 13,
                                      "value" : dt[1]
                               } , ignore_index=True )
    pkid = count
    
print("-----------------------------")
print("Data Retrieval Complete for Consumption")
print("-----------------------------")
```
