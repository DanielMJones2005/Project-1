```python
#http://api.eia.gov/series/?api_key=YOUR_API_KEY_HERE&series_id=TOTAL.TXEIEUS.A
url = "http://api.eia.gov/series/?api_key="
api_code1="&series_id=TOTAL."
api_code2=".A"
category=["CLEIEUS","DKEIEUS","GEEIEUS","NNEIEUS","NWEIEUS",
         "PAEIEUS","PCEIEUS","RFEIEUS"
         ]

#Total Category= "TXEIEUS"
#query_url=f"{url}{api_key}&series_id=TOTAL.TXEIEUS.A"
#print(query_url)

#eia_json = requests.get(query_url).json()
#http://api.eia.gov/series/?api_key=d7c7d5a92f23e68e5710fc5cfbd273dc&series_id=TOTAL.TXEIEUS.A
#pprint(eia_json)
```

```python
start = []
end = []
f = []
name = []
sector = []
series_id = []
units = []
updated = []
year = []
emission = []
data = []
count_cat = 0

for cat in category:
        query_url = url + api_key + api_code1 + cat + api_code2
        count_cat = count_cat + 1
        print("Processing Record " + str(count_cat) + " | " + cat)

        emiss_response = requests.get(query_url).json()
        
       
        data.append(emiss_response["series"][0]["data"])
        ln = (len(data[0]))

        for i in range(0,ln):
            start.append(emiss_response["series"][0]["start"])
            end.append(emiss_response["series"][0]["end"])
            f.append(emiss_response["series"][0]["f"])
            name.append(emiss_response["series"][0]["name"])
            sector.append("Electric Power")
            series_id.append(emiss_response["series"][0]["series_id"])
            units.append(emiss_response["series"][0]["units"])
            updated.append(emiss_response["series"][0]["updated"])
            year.append(emiss_response["series"][0]["data"][i][0])
            emission.append(emiss_response["series"][0]["data"][i][1])
print("-----------------------------")
print("Data Retrieval Complete")
print("-----------------------------")
```

```python
emiss_elect_consumption_df = pd.DataFrame({
   "start":start,
   "end":end,
   "name": name,
   "sector": sector,
   "series_id": series_id,
   "units": units,
   "updated": updated,
   "year": year,
   "emission": emission
})
emiss_elect_consumption_df
```
