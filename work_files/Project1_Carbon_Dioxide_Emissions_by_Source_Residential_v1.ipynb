```python
#Project 1: EIA Data Sets > Total Energy > Environment > Carbon Dioxide Emissions From Energy Consumption: Residential Sector

# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import json
from tqdm import tqdm_notebook
from datetime import datetime
from pprint import pprint

# Import API key
from api_keys_res import api_key
```

```python
#http://api.eia.gov/series/?api_key=YOUR_API_KEY_HERE&series_id=TOTAL.CLRCEUS.A
#http://api.eia.gov/series/?api_key=YOUR_API_KEY_HERE&series_id=TOTAL.DFRCEUS.A
#http://api.eia.gov/series/?api_key=YOUR_API_KEY_HERE&series_id=TOTAL.HLRCEUS.A
#http://api.eia.gov/series/?api_key=YOUR_API_KEY_HERE&series_id=TOTAL.KSRCEUS.A
#http://api.eia.gov/series/?api_key=YOUR_API_KEY_HERE&series_id=TOTAL.NNRCEUS.A
#http://api.eia.gov/series/?api_key=YOUR_API_KEY_HERE&series_id=TOTAL.PARCEUS.A
#http://api.eia.gov/series/?api_key=YOUR_API_KEY_HERE&series_id=TOTAL.ESRCEUS.A
#http://api.eia.gov/series/?api_key=YOUR_API_KEY_HERE&series_id=TOTAL.TERCEUS.A



url = "http://api.eia.gov/series/?api_key="
api_code1 = "&series_id=TOTAL."
#api_energy_source = "TERCBUS"
api_period = ".A"    


energy_source = [
    "Coal",
    "Distillate Fuel Oil",
    "HGL",
    "Kerosene",
    "Natural Gas, Excluding Supplemental Gaseous Fuel",
    "Petroleum",
    "Residential Share of Electric Power",
    "Total Energy",
]


energy_source_code = [

    "CLRCEUS",
    "DFRCEUS",
    "HLRCEUS",
    "KSRCEUS",
    "NNRCEUS",
    "PARCEUS",
    "ESRCEUS",
    "TERCEUS",
]

energy_source_df = pd.DataFrame({
    "code": energy_source_code,  
    "source": energy_source,
      
})
energy_source_df
```

```python
start = []
end = []
f = []
name = []
sector = []
sec = "Residental"
code = []
source = []
series_id = []
units = []
updated = []
year = []
value = []
count_st = 0


for energy in energy_source_df["code"]:
    query_url = url + api_key + api_code1 + energy + api_period
    consume_res_resp = requests.get(query_url).json()

    print("Processing " + energy)
    
    data = []
    data.append(consume_res_resp["series"][0]["data"])
    ln = (len(data[0]))

    for i in range(0, ln):

        start.append(consume_res_resp["series"][0]["start"])
        end.append(consume_res_resp["series"][0]["end"])
        f.append(consume_res_resp["series"][0]["f"])
        name.append(consume_res_resp["series"][0]["name"])
        code.append(consume_res_resp["series"][0]["series_id"].split(".")[1])
        source.append("")
        sector.append(sec)
        series_id.append(consume_res_resp["series"][0]["series_id"])
        units.append(consume_res_resp["series"][0]["units"])
        updated.append(consume_res_resp["series"][0]["updated"])
        year.append(consume_res_resp["series"][0]["data"][i][0])
        value.append(consume_res_resp["series"][0]["data"][i][1])
        
print("-----------------------------")
print("Data Retrieval Complete")      
print("-----------------------------")
```

```python
#Create DataFrame
emissions_by_source_data = pd.DataFrame({
    "start": start,
    "end": end,
    "f": f,
    "name": name,
    "code": code,
    #"source": source,
    "sector": sector,
    "series_id": series_id,
    "units": units,
    "updated": updated,
    "year": year,
    "value": value
    
})
emissions_by_source_data
```

```python
emissions_by_source_df = pd.merge(emissions_by_source_data, energy_source_df, on="code")
```
