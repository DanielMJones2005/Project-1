```python
#Project 1: EIA Data Sets > CO2 Emissions > by Fuel > All fuels > Residential carbon dioxide emissions by state

# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import json
from tqdm import tqdm_notebook
from datetime import datetime
from pprint import pprint

# Import API key
from api_keys_res import api_key
```

```python
url = "http://api.eia.gov/series/?api_key="
api_code1 = "&series_id=EMISS.CO2-TOTV-RC-TO-"
api_code2 = ".A"    

state_abb = ["AL", "AK", "AZ", "AR", "CA", "CO", "CT", "DE", "FL", "GA", 
            "HI", "ID", "IL", "IN", "IA", "KS", "KY", "LA", "ME", "MD",
            "MA", "MI", "MN", "MS", "MO", "MT", "NE", "NV", "NH", "NJ",
            "NM", "NY", "NC", "ND", "OH", "OK", "OR", "PA", "RI", "SC",
            "SD", "TN", "TX", "UT", "VT", "VA", "WA", "WV", "WI", "WY"]
```

```python
start = []
end = []
f = []
geo = []
state = []
name = []
sector = []
series_id = []
source = []
units = []
units_short = []
updated = []
year = []
emission = []
count_st = 0

for st in state_abb:
    query_url = url + api_key + api_code1 + st + api_code2
    count_st = count_st + 1
    print("Processing Record " + str(count_st) + " | " + st)
    
    emiss_resi_resp = requests.get(query_url).json()
    
    data = []
    data.append(emiss_resi_resp["series"][0]["data"])
    ln = (len(data[0]))
    

    #length = int(total)
    for i in range(0, ln):
        start.append(emiss_resi_resp["series"][0]["start"])
        end.append(emiss_resi_resp["series"][0]["end"])
        f.append(emiss_resi_resp["series"][0]["f"])
        geo.append(emiss_resi_resp["series"][0]["geography"])
        state.append(emiss_resi_resp["series"][0]["geography"].split("-")[1])
        name.append(emiss_resi_resp["series"][0]["name"])
        sector.append(emiss_resi_resp["series"][0]["name"].split(" ")[0])
        series_id.append(emiss_resi_resp["series"][0]["series_id"])
        source.append(emiss_resi_resp["series"][0]["source"])
        units.append(emiss_resi_resp["series"][0]["units"])
        units_short.append(emiss_resi_resp["series"][0]["unitsshort"])
        updated.append(emiss_resi_resp["series"][0]["updated"])
        year.append(emiss_resi_resp["series"][0]["data"][i][0])
        emission.append(emiss_resi_resp["series"][0]["data"][i][1])

print("-----------------------------")
print("Data Retrieval Complete")      
print("-----------------------------")
```

```python
#Create DataFrame
emiss_res_df = pd.DataFrame({
    "start": start,
    "end": end,
    "f": f,
    "geo": geo,
    "state": state,
    "name": name,
    "sector": sector,
    "series_id": series_id,
    "source": source,
    "units": units,
    "units_short": units_short,
    "updated": updated,
    "year": year,
    "emission": emission
    
})
```
